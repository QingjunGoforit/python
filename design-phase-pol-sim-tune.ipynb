```python
#simulation
#phase noise on arm2
import numpy as np
#from scipy import signal
import matplotlib.pyplot as plt
amplitude_sine1=(1/4)*np.pi
amplitude_sine2=(3/2)*np.pi
time_range=20
period_sine=1
num_period_sine=int(time_range/period_sine)
samples_per_period=500

time = np.linspace(0, num_period_sine*period_sine, num_period_sine*samples_per_period, endpoint=False)

phase_noise1=amplitude_sine1*(np.sin(time)+1) #phase
phase_noise2=amplitude_sine2*(np.sin(2*(time+np.pi/4))+1) #polarization
phase_noise_diff=phase_noise2-phase_noise1

size=20
size1=15
figHandler, axHandler = plt.subplots(1, 2, figsize=(10, 5),layout='constrained')
axHandler[0].plot(time, phase_noise1/np.pi, label='V1: modulation for phase')
axHandler[0].plot(time, phase_noise2/np.pi, 'g',label='V2: modulation for polarization')
axHandler[0].set_xlabel('time',fontsize=size)
axHandler[0].set_ylabel('Vpp in Vpi',fontsize=size)
axHandler[0].legend()
axHandler[0].tick_params(axis='both', labelsize=size)  # Set font size for ticks

axHandler[1].plot(time, phase_noise_diff/np.pi, label='phase difference on two arm')
axHandler[1].set_xlabel('time',fontsize=size)
axHandler[1].set_ylabel('phase in pi',fontsize=size)
axHandler[1].legend()
axHandler[1].tick_params(axis='both', labelsize=size1)  # Set font size for ticks


print("number of dots in sine wave",num_period_sine*samples_per_period)
```
